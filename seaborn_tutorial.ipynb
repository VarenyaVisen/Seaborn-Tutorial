### Seaborn
* Data visualization library

```python
# importing library
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

Note: Seaborn has some built in data set

```python
# total bill vs tip data set
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Setting the theme for the plots
sns.set_theme()
```

```python
# visualize the data
sns.relplot(data=tips, x='total_bill', y='tip', col='time', hue='smoker',style='smoker',size='size')
```

```python
#House price datset
from sklearn.datasets import fetch_california_housing
california_dataset = fetch_california_housing()
california_df = pd.DataFrame(california_dataset.data, columns = california_dataset.feature_names)
type(california_df)
```

```python
california_df.head()
```

Distribution plot

```python
sns.displot(california_df['HouseAge'])
```

Correlation
* Positive : a positive correlation means that two variables move in the same direction (both increase or both decrease)
* Negative : a negative correlation means that two variables move in opposite directions (as one increases, the other decreases)

Heat Map

```python
correlation = california_df.corr()
```

```python
# constructing a heat map
plt.figure(figsize=(10,10))
sns.heatmap(correlation,cbar=True, square=True, fmt='.1f', annot=True, annot_kws={'size' : 8},cmap='Blues')
```
